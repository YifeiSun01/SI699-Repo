```python
import pandas as pd
import numpy as np
from google.colab import drive

drive.mount('/content/drive')
psrc_trips = pd.read_csv("psrc_trips.csv")
psrc_trips.columns
```

```python
grouping = psrc_trips.groupby(["o_tract10","d_tract10"]).count().reset_index()
origin = grouping["o_tract10"].values
dest = grouping["d_tract10"].values
weight = grouping["household_id"].values
```

```python
import networkx as nx
import numpy as np

G = nx.DiGraph()

# Add edges with weights
for i in range(len(origin)):
    G.add_edge(origin[i], dest[i], weight=weight[i])
```
